```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

%matplotlib inline

import os
print(os.listdir())

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv(r'C:\Users\anmol\Downloads\heart (1).csv' , encoding = 'ISO-8859-1')
```

```python
#CHECKING THE DISTRIBUTION OF TARGET VARIABLE 
# 1 represents defective heart
# 0 represents healthy heart
dataset['target'].value_counts()
```

```python
# SPLITTING TARGET AND OTHER FEATURES 
#axis=0: Refers to rows (vertical axis).
#axis=1: Refers to columns (horizontal axis).

X = dataset.drop(columns = 'target' , axis =1)
Y = dataset['target']
print(X)
print(Y)
```

```python
#SPLITTING DATA INTO TRAINING AND TEST DATA
#stratify=Y parameter ensures that the distribution of the target variable (Y) is preserved in both the training and testing datasets.

X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , stratify = Y , random_state = 2)
```

```python
#DATA POINTS IN THE DISTRIBUTION
print(X.shape , X_train.shape , X_test.shape)
```

```python
#MODEL TRAINING 
#LOGISTIC REGRESSION MODEL

model = LogisticRegression()

#TRAINIG 
model.fit(X_train , Y_train)
```

```python
#MODEL EVALUATION 
# ACCURACY SCORE AS A EVALUATION METRICS

#ACCURACY ON TRAINING DATA 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train) 
```

```python
print('ACCURACY ON TRAINING DATA : ' , training_data_accuracy*100 , '%')
```

```python
#ACCURACY ON TEST DATA
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(X_test_prediction , Y_test) 
```

```python
print('ACCURACY ON TEST DATA : ' , training_data_accuracy*100 , '%')
```
